```python
import spacy
from spacy import displacy
```

```python
# example of model usage
def render(text):
  
    nlp = spacy.load('en_core_web_sm') # model
  
    doc = nlp(text) # data processing

    # FYI all the properties
    props = [p for p in dir(doc.ents[0]) if p[0] != '_']
    print(props)

    # custom processing of the answer

    # get counts of entities
    ent_labels = [e.label_ for e in doc.ents]
    freq = dict()
    for l in ent_labels:
        freq[l] = ent_labels.count(l)
    print(freq)

    # get coordinates of entities
    coordinates = [e.start_char for e in doc.ents]
    print(coordinates)

    # render the answer
    displacy.render(doc, style='ent', jupyter=True)
```


### Пример оформления инварианта

Рассмотрим задачу поиска подстроки в строке. 

Предполагаем, что алгоритм должен находить все вхождения подстроки.

> **MR1.** Если в строке S найдёна некоторая подстрока s ровно k раз, то в строке SS она будет найдена не менее 2k раз (возможны нахождения на месте склейки строк).

Формально. Пусть S - строка, s - её подстрока, f(S,s) - определённое программой число вхождений s в S. Тогда f(SS,s) >= 2*f(S,s).


```python
import re
import unittest

# function for testing
def func_to_test(substr, string):
    return re.finditer(pattern=substr, string=string)


class TestStringMethods(unittest.TestCase):

    def test_with_oracle1(self):
        # input data
        big_string = 'abacaba'
        substr = 'aba'

        # correct answer
        right_answer = [0, 4]

        indices = func_to_test(substr, big_string)
        answer = [index.start() for index in indices]
        
        self.assertTrue(answer == right_answer)

    def test_metamorphic1(self):
        # input data
        big_string1 = 'abacab'
        big_string2 = big_string1 + big_string1
        substr = 'aba'

        # first answer
        indices = func_to_test(substr, big_string1)
        indices1 = [index.start() for index in indices]

        # second answer
        indices = func_to_test(substr, big_string2)
        indices2 = [index.start() for index in indices]

        # metamorphic relation
        self.assertTrue(2*len(indices1) <= len(indices2))
```

### Реализуйте собственные тестовые инварианты ниже

#### Тесты с оракулом

1. Тест, в котором на вход поступает заданный текст, ответ модели сравнивается с эталонным (`test_oracle_1`)

2. То же самое, что и первый, только вход генерируется с помощью повторения образца. Ответ также получается умножением результатов для образца. Новых вхождений не появляется (из-за формата конкатенации и вида образца) (`test_oracle_2`)

#### Инварианты

1. Тест основан на теории множеств: модель извлекает из текста ключевые слова, следовательно, запуск модели на тексте $B$, являющимся перестановкой слов текста $A$ должен вернуть тот же самый результат. Тут под словом "слово" я подразумеваю один объект, например, город или компания, состоящие из нескольких слов являются одним объектом (`test_shuffle_invariant`)

$$
f(x) = \text{model}.predict(\text{text}) \\
R(x, y, f(x), f(y)) = [y = \text{shuffle words}(x)] \rightarrow [f(x) = f(y)]
$$

2. Тест основан на теории множеств: модель извлекает из текста ключевые слова, следовательно, запуск модели на тексте $B$, содержащим в себе текст $A$ как подстроку, может только расширить множество найденных ключевых слов (`test_add_invariant`)

$$
f(x) = \text{model}.predict(\text{text}) \\
R(x, x + y, f(x), f(x + y)) = [f(x) \subseteq f(x + y)]
$$

3. При повторном запуске модель должна давать такие же результаты (`test_retry_invariant`)

$$
f(x) = \text{model}.predict(\text{text}) \\
R(x, y, f(x), f(y)) = [y = x] \rightarrow [f(x) = f(y)]
$$

4. Тест основан на теории множеств: модель извлекает из текста ключевые слова, следовательно, запуск модели на тексте $B$, полученном из текста $A$ удалением некоторых случайных слов, не может добавить новых ключевых слов (`test_delete_invariant`)

$$
f(x) = \text{model}.predict(\text{text}) \\
R(x, y, f(x), f(y)) = [y \subseteq x] \rightarrow [f(y) \subseteq f(x)]
$$

```python
from random import shuffle


nlp = spacy.load("en_core_web_sm") 
print(nlp.get_pipe("ner").labels)


text = "Google is American company. Google is American company. Google is American company. Google is American company"
doc = nlp(text)

displacy.render(doc, style='ent', jupyter=True)
```

```python
from dataclasses import dataclass
from collections import defaultdict
import random


@dataclass
class Statistic:
    cnt: int
    coordinates: list[tuple[int, int]]

class TestModel(unittest.TestCase):
	nlp = spacy.load("en_core_web_sm")

	def render(self, text: str) -> dict[str, Statistic]:
		doc = self.nlp(text)

		entries = defaultdict[str, list[tuple[int, int]]](list)
		for ent in doc.ents:
			entries[ent.label_].append((ent.start_char, ent.end_char))

		stats = {label: Statistic(cnt=len(coords), coordinates=coords) for label, coords in entries.items()}

		return stats
  
	def test_oracle_1(self):
		# Yandex не распознается как компания
		text = "Sber is a Russian company with its head office in Moscow. Google is an American company, the head office is located in MountainView"
		stats = self.render(text)

		self.assertEqual(stats["ORG"].cnt, 2)
		self.assertEqual(stats["NORP"].cnt, 2)
		self.assertEqual(stats["GPE"].cnt, 2)

		for label in nlp.get_pipe("ner").labels:
			if label not in ("ORG", "NORP", "GPE"):
				self.assertTrue(label not in stats.keys())

	def test_oracle_2(self):
		text = " ".join(["Google is American company."] * 100)
		stats = self.render(text)

		self.assertEqual(stats["ORG"].cnt, 100)
		self.assertEqual(stats["NORP"].cnt, 100)

		for label in nlp.get_pipe("ner").labels:
			if label not in ("ORG", "NORP"):
				self.assertTrue(label not in stats.keys())

	def test_shuffle_invariant(self):
		def invariant(text: str):
			original_stats = self.render(text)
	
			words = text.split(" ")
			random.shuffle(words)
			text = " ".join(words)
			
			modified_stats = self.render(text)

			self.assertSetEqual(set(original_stats.keys()), set(modified_stats.keys()))

			for label in original_stats.keys():
				self.assertEqual(original_stats[label].cnt, modified_stats[label].cnt)

		text = "Toyota is a Japanese company with its head office in Tokyo. Google is an American company, the head office is located in Washington"
		invariant(text)

	def test_add_invariant(self):
		def invariant(text_1: str, text_2: str):
			stats_1 = self.render(text_1)
			stats_2 = self.render(text_2)
			concat_stats = self.render(text_1 + " " + text_2)

			for key in stats_1.keys():
				self.assertIn(key, concat_stats.keys())
				self.assertGreaterEqual(concat_stats[key].cnt, stats_1[key].cnt)

			for key in stats_2.keys():
				self.assertIn(key, concat_stats.keys())
				self.assertGreaterEqual(concat_stats[key].cnt, stats_2[key].cnt)

		text_1 = "Toyota is a Japanese company with its head office in Tokyo. Google is an American company, the head office is located in Washington"
		text_2 = "Toyota is not a Russian company with its head office not in Moscow. Google is an American company, the head office is located in MountainView"
		invariant(text_1, text_2)

	def test_retry_invariant(self):
		def invariant(text: str):
			stats_1 = self.render(text)
			stats_2 = self.render(text)
			stats_3 = self.render(text)

			self.assertEqual(stats_1, stats_2)
			self.assertEqual(stats_2, stats_3)

		text = "Toyota is a Japanese company with its head office in Tokyo. Google is an American company, the head office is located in Washington"
		invariant(text)

	def test_delete_invariant(self):
		def invariant(text: str):
			original_stats = self.render(text)

			words = text.split(" ")
			new_words = []
			for word in words:
				if random.random() > 0.8:
					new_words.append(word)
				
			new_text = " ".join(new_words)
			modified_stats = self.render(new_text)

			for label in modified_stats.keys():
				self.assertIn(label, original_stats.keys())
				self.assertLessEqual(modified_stats[label].cnt, original_stats[label].cnt)

		text = "Toyota is a Japanese company with its head office in Tokyo. Google is an American company, the head office is located in Washington"
		invariant(text)
```

```python
unittest.TextTestRunner(verbosity=2).run(
    unittest.TestLoader().loadTestsFromTestCase(TestModel)
)
```

#### Сравнение методов.

Каждый метод имеет свои преимущества и недостатки.

- Тестирование с оракулом - зачастую прост в реализации (например, при проверке детерминированного кода), легко интерпретируется, однако в некоторых случаях может быть очень дорогим (например, в ML - ручная разметка выборки или даже невозможным - компиляторы)

- Тестирование инвариантами - позволяет тестировать достаточно сложные вещи (например, работу ML модели), при наличии инварианта достаточно дешевая реализация, высокая эффективность из-за наличия очень большого количества тестовых примеров (зачастую входные данные можно генерировать в практически бесконечном объеме), однако требует ручного подбора хорошего инварианта, сложно интерпретируется, может быть нестабильным

В целом зачастую для тестирования обычных программ почти всегда хватает тестирования с оракулом (юнит тесты), для тестирования же более сложных и абстрактных вещей, когда тестирование с оракулом не подходит, подойдет тестирование инвариантами.
